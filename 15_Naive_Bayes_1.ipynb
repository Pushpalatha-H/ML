#Naive Bayes

Gaussian Naive Bayesis used is a classification technique used in ML based on probablistic approach and Gaussian Distribution

P(A/B) = P(B/A) * P(A)/ P(B)

Predict Titanic survival using naive Bayes

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/14_naive_bayes/titanic.csv")
df.head()
```

```python
df =df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis='columns')
df.head(3)
```

```python
# check if it has n.a
df.isna().sum()
```

```python
median = df['Age'].median()
```

```python
#  fill n.a. by median
df['Age']= df['Age'].fillna(median)
```

```python
dummies = pd.get_dummies(df['Sex']).astype('int')
dummies.head(2)
```

```python
final = pd.concat([df, dummies], axis='columns')
final.head()
```

```python
x = df.drop(['Survived', 'Sex'], axis='columns')
x.head(2)
```

```python
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
# model building
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.predict(x_test)
```

```python
model.predict_proba(x_test[0:10])
```

#cross_val_score

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(GaussianNB(), x_train, y_train, cv=5)
```
