# RandomForestClassifier
RandomForestClassifier is another popular modelused for both Regression and Classification

Forest means Trees in MachineLearning tree means DecisionTree

How to build multiple decision trees out of a single data set?

Take a dataset build a subsets and then build decision tree.

Since we did random samples here it is RandomForest and we have multiple trees.

Now all the trees are formed once. it, is trained to take majority of vote of all Decision

This is the basic criteria behind random Forest Algorithm

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data
```

```python
digits.target
```

```python
plt.matshow(digits.images[0])
```

```python
plt.gray()
for i in range(5):
  plt.matshow(digits.images[i])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

**RandomForestClassifier Model**

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=40)
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
```

**Confusion_Matrix**

Used to find where my Model is performing better and where it is dooing poor

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# heatmap
import seaborn as sn
plt.figure(figsize= (10,7))
sn.heatmap(cm, annot = True)
plt.xlabel('Prediction')
plt.ylabel('Truth')
```
