# Problem to predict home price

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/homeprices.csv")
df.head()
```

```python
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.title("Area V/s Price")
```

```python
x = df.drop('price', axis = 'columns')
x
```

```python
y = df['price']
y
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[3000]])
```

```python
model.predict([[1200]])
```

```python
area_df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/areas.csv")
area_df
```

```python
p = model.predict(area_df)
```

```python
p
```

```python
area_df['price'] = p
```

```python
area_df
```

```python
area_df.to_csv('prediction.csv')
```

##Excercise

Predict canada's per capita income in year 2020. There is an exercise folder here on github at same level as this notebook, download that and you will find canada_per_capita_income.csv file. Using this build a regression model and predict the per capita income fo canadian citizens in year 2020

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv")
df.head()
```

```python
x = df.drop('per capita income (US$)', axis = 'columns')
y = df['per capita income (US$)']
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()

model.fit(x,y)
```

```python
model.score(x,y)
```

#per capita income for the year 2020

```python
model.predict([[2020]])
```
