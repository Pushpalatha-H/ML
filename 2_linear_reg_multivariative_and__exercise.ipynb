#ML with python: Linear Regression Multiple Variables

Sample problem of predicting home price in monroe, new jersey (USA)
Below is the table containing home prices in monroe twp, NJ. Here price depends on area (square feet), bed rooms and age of the home (in years). Given these prices we have to predict prices of new homes based on area, bed rooms and age.


Given these home prices find out price of a home that has,

3000 sqr ft area, 3 bedrooms, 40 year old

2500 sqr ft area, 4 bedrooms, 5 year old

```python
# importing pre-requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/2_linear_reg_multivariate/homeprices.csv")
df
```

```python
median = df['bedrooms'].median()
```

```python
df['bedrooms'] = df['bedrooms'].fillna(median)
```

```python
df
```

```python
x = df.drop('price', axis = 'columns')
y = df['price']
```

```python
x
```

```python
y
```

# Build model using linear_regression

```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score
```

```python
model.predict([[3000,3,15]])
```

```python
model.predict([[1200, 2, 0]])
```

#Exercise

In exercise folder (same level as this notebook on github) there is hiring.csv. This file contains hiring statics for a firm such as experience of candidate, his written test score and personal interview score. Based on these 3 factors, HR will decide the salary. Given this data, you need to build a machine learning model for HR department that can help them decide salaries for future candidates. Using this predict salaries for following candidates,

2 yr experience, 9 test score, 6 interview score

12 yr experience, 10 test score, 10 interview score



```python
import pandas as pd
import numpy as np
!pip install word2number
from word2number import w2n
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/2_linear_reg_multivariate/Exercise/hiring.csv")
df
```

```python
df['experience'] = df['experience'].fillna('zero')
```

```python
mean = df['test_score(out of 10)'].mean()
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(mean)
```

```python
df
```

```python
# convert df['experience'] to numbers using w2n.word_to_number
df['experience'] = df['experience'].apply(w2n.word_to_num)
```

```python
df
```

```python
x = df.drop('salary($)', axis ='columns')
x.head()
```

```python
y = df['salary($)']
y.head()
```

```python
# model building
from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
model.fit(x, y)
```

```python
model.score(x,y)
```

2 yr experience, 9 test score, 6 interview score



```python
model.predict([[2,9,6]])
```

12 yr experience, 10 test score, 10 interview score

```python
model.predict([[12,10,10]])
```
