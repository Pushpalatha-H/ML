# Decision Tree Classifier
If we need to classify further, further,.. to predict out then we can use DecisionTree Classifier

**We need to find out whether a persons salary is more than 100k, based on company, job and degree**

```python
import pandas as pd   #import pre requests
import numpy as np
```

```python
# read csv
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/salaries.csv")
df.head()
```

```python
# ML only takes number as input so we should do label encoding for the columns
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['le_job'] = le.fit_transform(df['job'])
df['le_degree'] = le.fit_transform(df['degree'])
df['le_company'] = le.fit_transform(df['company'])
```

```python
df.head()
```

```python
x = df.drop(['company', 'job', 'degree', 'salary_more_then_100k'], axis = 'columns')
y = df['salary_more_then_100k']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
# DecisionTree Model
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

Is Computer Engineer, Bachelors degree working in Google has salary > 100 k ?

```python
model.predict([[1,0,2]])
```
